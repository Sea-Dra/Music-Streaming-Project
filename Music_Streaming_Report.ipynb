```python
genres_v2.rename(columns={'duration_ms': 'duration'}, inplace=True) # Changing the name of the column
```

```python
def format_duration(minutes):
    mins = int(minutes)  # Get the whole minutes
    secs = int((minutes % 1) * 60)  # Calculate the seconds
    return f"{mins}:{secs:02d}"  # Format as 'minutes:seconds'

# Apply the formatting function to the 'duration' column
genres_v2['duration'] = genres_v2['duration'].apply(format_duration)



```

```python
genres_v2
```

```python
sns.scatterplot(x='danceability', y='tempo',hue='tempo', palette='viridis', data=genres_v2)
plt.show()
```

```python

sns.lineplot(x='tempo', y='danceability', data= genres_v2)
plt.title('Correlation between Tempo and Danceability')
plt.xlabel('Tempo')
plt.ylabel('Danceability')
plt.show()
```

1) What is the correlation between Tempo and Danceability?

There is no correlation between them. In both cases, there will be music with low and high tempos that are danceable.
However, you will notice music with the lowest tempos has less music than the rest.
I used a scatterplot first because it is the best when comparing two Numerical columns

```python
genre_acousticness = genres_v2.groupby('genre')['acousticness'].mean().reset_index()


sns.barplot(x='genre', y='acousticness', data=genres_v2)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis label so they don't overlap
plt.title('Average Acousticness by Genre')
plt.show()
```

```python
pivot_table = pd.pivot_table(genres_v2, values='acousticness', index='genre', aggfunc='mean')

sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title('Average Acousticness by Genre')
plt.show()
```

2) Does the acousticness differ based on the genre?

The answer is yes. As we can see in both charts, the heatmap and the bar chart. 
The genre with the highest acousticness levels is 'RnB', Hip hop falls second to that.
Just to let you know, this table does not have country music. All the genres are in the charts above.
However, Rnb takes the lead in the genres this table contains.
The lowest on the list is Psytrance.
