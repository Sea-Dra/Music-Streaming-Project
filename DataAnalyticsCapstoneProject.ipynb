## Question 3: Key vs Liveness

```python
plt.figure(figsize=(10,6))
sns.histplot(data=genres_v2,x='key',y='liveness')
plt.title('Distribution Between Musical Key and Liveness of the Song',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix=genres_v2[["key","liveness"]].corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title('Relationship Between Musical Key and Liveness of the Song',fontsize=14)
plt.show()
```

For Question 3, we aimed to show the distribution of the key and liveness variables and their correlation. To achieve this, we initially did a scatterplot to visualize the distribution of the key and liveness variables, but the scatterplot showed the plot points in a vertical line and then in a horizontal line when we switched the x and y axis. So, we then did a histogram to visualize the distribution. Looking at the histogram, it seems that the majority of the liveness of the songs is below 0.2. This means there is little to no detection of an audience present when the songs were recorded. To display the correlation between key and liveness, we first calculated the correlation of the key and liveness variable and then we displayed our results using a heatmap. Looking at the heatmap, there is a very weak positive correlation between key and liveness. This suggests that the musical key used to record songs has nothing to do with whether or not there's an audience present when a song is being recorded.

## Question 4: Most Popular Genres for playlists.csv

```python
playlists_counts=playlists['Genre'].value_counts().reset_index()
playlists_counts.columns=['Genre', 'Count']
plt.figure(figsize=(10,6))
sns.barplot(data=playlists_counts,x='Genre',y='Count')
plt.title('Count of Genres in playlists.csv',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
total = playlists['Genre'].value_counts().values.sum()

def fmt(x):
    return '{:.1f}%\n{:.0f}'.format(x, total*x/100)

plt.pie(playlists['Genre'].value_counts().values, labels=playlists['Genre'].value_counts().index, autopct=fmt)
ax.set_title('Share of Playlists by Genre')
plt.show()
```

For Question 4, we aimed to show how popular certain genres were in the playlists.csv. To achieve this, we did a bar chart and a pie chart. The bar chart showed that Emo is the most popular genre. The pie chart further breaks down the count of how many playlists there were in the dataset by genre and the percent share that they take up in the dataset.
